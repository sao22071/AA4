Averigua la suma, la media y la desviación de los puntos marcados por los Boston Celtics (BOS) en toda su historia

```python
# Variable filtro
filtro = nba['team_id'] == 'BOS'

boston_Celtics = nba.loc[filtro]

print(f"La suma de los puntos marcados: {boston_Celtics['pts'].sum()}")
print(f"La media de los puntos marcados: {boston_Celtics['pts'].mean()}")
print(f"La desviación estándar de los puntos marcados: {boston_Celtics['pts'].std()}")
```

Calcula el total de años que llevan jungando los Celtics. Y luego calcula el promedio de puntos por año


```python

#promedio de puntos por año
#se convierte año a fecha y se crea una nueva columna
nba['date_played'] = pd.to_datetime(nba['date_game'])

#se crea la columna año extrayendo el año de date_played
nba['año'] = nba['date_played'].dt.year

#se vuelve a crear el dataset para que tome los valores de las nuevas columnas
boston_Celtics = nba[nba['team_id'] == 'BOS']

# se agrupa el DataFrame filtado por año y a cada año se le calcula el promedio de los puntos
# Ahora filtra después de crear la columna 'año'
promedio_punto_año = boston_Celtics.groupby('año')['pts'].mean()
print(f'Promedio de puntos por año\n {promedio_punto_año}')
```


Ahora busca el total de años jugados por los Celtics. Usa la fución `apply` y una función lambda

```python
# #Cantidad total de años que llevan jugando
cantidad_años = boston_Celtics[['año']].apply(lambda x: x.max() - x.min() + 1)
años_jugados = cantidad_años['año']
print(f'Cantidad años jugados por los Celtics: {años_jugados}')
```

Ahora que calculaste el numero total de años jugados, calcula el total de puntos por año

```python
total_puntos_año = boston_Celtics['pts'].sum()
print(f'Promedio de puntos por año\n {total_puntos_año/años_jugados}')
```

Realiza un gráfico de dispersión de los puntos de los knicks vs los puntos de sus oponentes en los años arriba del año 2000

```python
#Selecciona los puntos de los Nicks
des_knicks_2000 = nba_copy[(nba_copy['team_id'] == 'NYK') & (nba_copy['year_id'] > 2000)]['pts']

#Seleccion los puntos de los oponentes
des_opp_2000 = nba_copy[(nba_copy['team_id'] == 'NYK') & (nba_copy['year_id'] > 2000)]['opp_pts']
```

Encuenta el número de juegos ganados y perdids por los Miami heats en el año 2013

```python
# Juegos Ganados
juegos_ganados = nba_copy[(nba_copy['team_id'] == "MIA")
                        & (nba_copy['fran_id']=="Heat")
                        & (nba_copy['year_id'] == 2013)
                        & (nba_copy['pts'] > nba_copy['opp_pts'])]

# Juegos perdidos
juegos_perdidos = nba_copy[(nba_copy['team_id'] == "MIA")
                        & (nba_copy['fran_id']=="Heat")
                        & (nba_copy['year_id'] == 2013)
                        & (nba_copy['pts'] < nba_copy['opp_pts'])]

print(f'W: {len(juegos_ganados)}\nP: {len(juegos_perdidos)}')
```

Grafica tu resultado con un grafico de torta
